```python
import json
import time
import requests
import numpy as np
import pandas as pd
from io import StringIO
```

```python
apikey = "input API Key"

API_URL = "https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&apikey=" + apikey + "&symbol="

energy = ['XOM', 'CVX', 'COP', 'KMI','PSX','MPC', 'EPD', 'SLB', 'MPLX', 'VLO', 'EOG']
tech = ['T', 'VZ', 'ZM', 'ADBE','MSFT','ORCL', 'BKI', 'MDB', 'VRSN', 'SQ', 'AAPL', 'INTC', 'CRM', 'DOCU']
```

```python
response_json = []
```

```python
for x in energy:
    post_response = requests.get(API_URL + x).json()
    response_json.append(post_response)
    print(f"We have {len(response_json)} items")
    time.sleep(10)
```
